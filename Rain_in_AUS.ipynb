- checking the null count for each row

```python
df.isnull().sum().sort_values(ascending=False)
```

```python
check = df.isnull().sum().sort_values(ascending=False)
```

```python
check
```

```python
zero_pct = (df.isnull().sum() / df.isnull().count()).sort_values(ascending=False)
```

```python
missing = pd.concat([check,zero_pct],axis=1 , keys = ['amount','pct'])
```

```python
missing
```

# delete 'Sunshine' , ' Evaporation , 'Clould3pm' , 'Cloud9am'

- selected 4 features that has too much missing values

```python
unece_feat = list(missing[missing['pct']>0.15].index)
```

```python
unece_feat
```

```python
df_new = df.drop(unece_feat , axis=1)
```

```python
df_new = df_new.drop(['Date'],axis=1)
df_new = df_new.drop(['Location'],axis=1)
```
