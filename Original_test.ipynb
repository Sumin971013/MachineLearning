```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
from os.path import join
from sklearn.ensemble import IsolationForest
from sklearn.metrics import confusion_matrix
from sklearn.metrics import accuracy_score

from sklearn.metrics import roc_curve
from sklearn.metrics import roc_auc_score
import random
import warnings
warnings.filterwarnings('ignore')
```

```python
def draw_roc(test_label , test_score , num_of_features):
    
    
    roc_score = roc_auc_score(test_label,test_score)
    roc_score = round(roc_score,4)
    
    fpr,tpr,thresholds = roc_curve(test_label , test_score, pos_label=1)
    
    
    plt.figure(figsize=(7,7)) 
    
    
    
    plt.title('Number of features used : {}'.format(num_of_features) , fontsize=20)
    plt.plot(fpr,tpr,'k--' , lw=2 , label="auc: " + str(roc_score))
    plt.plot([0,1],[0,1])
    plt.legend(prop={"size":15})
    
    plt.show()
    


def select_features(data , test_data , num1,num2):
    
    return data.iloc[:, num1:num2] , test_data.iloc[:, num1:num2] 


def train_forest(train_data):
    
    #default model
    model = IsolationForest()
    #Learn
    model.fit(train_data)
    
#     test_predict = model.predict(test_data)
#     test_score = model.score_samples(test_data)
    
    return model
    
    
    
    
    
    
    
    

    
    
    
    
    
        
    
```

```python
def add_columns(data,test):
    
    columns = [f"w{i}" for i in range(len(data.columns)-1)]
    columns.append('is_normal')
    
    data.columns = columns
    test.columns = columns
    
    
    return data,test


def split_label(data,test):
    
    data,test = add_columns(data,test)
    
    train_data = data.iloc[:,:-1]
    train_label = data.iloc[:,[-1]]
    
    
    test_data = test.iloc[:,:-1]
    test_label = test.iloc[:,[-1]]
    
    test_label = test_label.replace(1,-1)
    test_label = test_label.replace(0,1)

    #train data 작용
    train_label = train_label.replace(0,1)
    
    
    return train_data, train_label , test_data, test_label


    


    
    
    
    


    
```

#  ROC & AUC for org 3

```python
df_3 = pd.read_csv(join("if_intern/org",'train_reuter_org_3.csv'),header=None)
test_3 = pd.read_csv(join("if_intern/org",'test_reuter_org_3.csv'),header=None)
```

```python
df_3 = pd.read_csv(join("if_intern/org",'train_reuter_org_3.csv'),header=None)
test_3 = pd.read_csv(join("if_intern/org",'test_reuter_org_3.csv'),header=None)
#split label
train_3 , train_label_3 , test_3, test_label_3 = split_label(df_3,test_3)

#choose number of feature to fit model
train_3_data , test_3_data = select_features(train_3,test_3,0,train_3.shape[1])

#fit
model3 = train_forest(train_3_data)

# get the test_score for model
test_score_3 = model3.score_samples(test_3_data)

#draw roc curve and auc
draw_roc(test_label_3 , test_score_3 , train_3.shape[1])
```
